## Scenario

As a junior data analyst, we are working with the marketing analyst team at Cyclistic. The task is to understand the key differences between how casual riders and annual members use Cyclistic bikes. With those insights, we will design a new marketing strategy to convert casual riders into annual members.

## Background

Cyclistic is a bike-share service in Chicago that was successfully launched in 2016 and now features more than 5800 bicycles and 600 docking stations. Along with standard two-wheeled bikes, the company also offers assistive options for bikers with disabilities such as reclining bikes and cargo bikes. Roughly 8% of bikers use the assistive option, and only 30% of bikers use this service to commute to work each day.

We are a part of the marketing analytics team responsible for collecting, analyzing, and reporting data related to all marketing campaigns and initiatives for the bike-share program. The director of marketing and our direct manager is Lily Moreno, who oversees all the team's activities. The primary audience will be the Cyclistic executive team, who will review and approve our recommended marketing programs for the company.

The current marketing strategy is to build upon general awareness and appeal to broad consumer segments. One successful approach was to offer flexibility in pricing plans: single-ride or full-day ride passes, and annual memberships. We classify customers with single-ride or full-day ride passes as casual riders and customers with annual memberships as annual members. 

The finance analysts have concluded that having annual members are much more profitable than having casual riders. Moreno believes maximizing annual members is the key to company growth so she suggests creating a marketing campaign that targets casual riders and converts them into annual members. In order to create this strategy, it is important for us to understand the following: what differs between casual riders and annual members, what prompts casual riders to purchase an annual membership, and how digital media affect our marketing efforts.

The data source used for this analysis comes from Lyft Bikes and Scooters, LLC (“Bikeshare”). The organization operates the Divvy bicycle sharing service located within the City of Chicago’s (“City”). Since Cyclistic is a fictitious company, we will be using data collected in the past twelve months by this organization as a basis to answer our business question [link] (https://divvy-tripdata.s3.amazonaws.com/index.html). This data is an appropriate representation of the customer data that would have been collected by a bikesharing service such as Cyclistic. Furthermore, the data has been made available for public use under the Divvy Data License Agreement [link] (https://ride.divvybikes.com/data-license-agreement).

The data sets used contain historical trip information of all riders that have accessed the Divvy bikeshare service between May 2021 to April 2022. The data consists of the type of bike taken, the trip start and end times, the trip start and end stations, and the type of customer. 

**Download the data**<br>
https://www.kaggle.com/datasets/kemenwong/divvy-trip-data

The datasets are organized into thirteen columns:<br>
**ride_id** - an unique identifier associated with a single ride from the moment the bike is checked out from the starting station to the moment it is checked into the ending station<br>
**rideable_type** - the type of bike that was borrowed and consists of three possible options: classic, docked, and electric<br>
**started_at** - the date and time (YYYY-MM-DD HH:MM:SS) that the bike was checked out<br>
**ended_at** - the date and time (YYYY-MM-DD HH:MM:SS) that the bike was checked in<br>
**start_station_name** - the specific station that the bike was checked out from<br>
**start_station_id** - the unique identifier given to each station<br>
**end_station_name** - denotes the specific station that the bike was checked into<br>
**end_station_id** - the unique identifier given to each station<br>
**start_lat** - the latitudinal coordinate associated with the starting station<br>
**start_lng** - the longitudinal coordinate associated with the starting station<br>
**end_lat** - the latitudinal coordinate associated with the ending station<br>
**end_lng** - the longitudinal coordinate associated with the ending station<br>
**member_casual** - the type of customer the rider is, either casual (single-ride or full-day passes) or member (annual memberships)


Based on the ROCCC system, there are no issues with the integrity or credibility of the data used. The data is reliable since it includes the processed information pertaining to each recorded ride that occurred within the past year, within the area of service. It has taken to account any rides that were taken by staff and any potentially false trips (false starts, re-docking bikes, etc.). The data is original since it is collected first-hand by Lyft Bikes and Scooters, LLC (“Bikeshare”). The data is  comprehensive in the scoop of the analysis but it is important to note that certain data such as personal identifiable information has been removed during anonymization as per privacy standards. This would lead to certain limitations within our analysis such as not being able to determine if customers reside within the area of service or account for casual customers purchasing multiple single-ride passes. The data is current as all datasets used within this analysis were from the past twelve months (May 2021 to April 2022). The data is credible since the Divvy system data in question is owned by the City of Chicago, which has reviewed and authorized the publication of its data for public use under its License Agreement. 

After assessing the licensing agreement, we are permitted to access, modify, and analyze the data as long as our use complies with the terms and conditions outlined within the document. The data was chosen because it can help us identify trends and relationships within and across the casual and member customer groups that use the bikeshare service. Insights regarding trip duration and frequencies could provide an explanation as to why customers remain casual riders and what could prompt them to become annual members.

```r
install.packages("ggplot2")
install.packages("ggrepel")
library(ggplot2)
library(ggrepel)
```

Since we generated an aggregated data set for analysis, we can begin to explore the variables of interest to see if there are any interesting trends and relationships. First, we explore the trip_duration variable which was calculated based on the difference (in mins) between the starting and ending times of a single recorded trip. The total trip durations (in mins) for each of the recorded months were added up using a function from the lubridate package.

```r
monthly_total_trip_duration <- complete_divvy_tripdata %>%
  group_by(month = lubridate::floor_date(started_at, "month")) %>%
  summarize(total_trip_duration = sum(trip_duration))

view(monthly_total_trip_duration)
```

From the resulting table, we notice that there is a greater total trip duration (in mins) in the summer months which becomes much lower during the winter season. We create a visualization of this to get a better understanding of just how drastic the change was. In order to do so, we appropriately convert the month variable from a date time data type to a date data type. Furthermore, we divide the total trip duration (in mins) by a million to get a better fit of values for the y-axis on the graph.

```r
monthly_total_trip_duration[1] <- as.Date(monthly_total_trip_duration$month)

monthly_total_trip_duration[c("total_trip_duration")] <- monthly_total_trip_duration[c("total_trip_duration")]/1e6 

ggplot(monthly_total_trip_duration, aes(x=month, y=total_trip_duration)) +
  geom_line() +
  geom_point() +
  scale_y_continuous(breaks = seq(0, 20, 2),
                     limits = c(0, 20)) +
  scale_x_date(date_breaks = "1 month", date_labels = "%b-%y") +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5)) +
  xlab("Month") +
  ylab("Total Trip Duration (mins/10^6)") +
  labs(title = "Total Trip Duration by Month", 
       subtitle = "Data from May 2021 - Apr 2022",
       caption = "Source: Divvy (https://ride.divvybikes.com/system-data)")

ggsave("figure_1.pdf")
```

As confirmed from our data table, we can see that ridership peaked during the summer months of June and July. It dropped considerably during the transition into fall months, and down to very low amounts during the winter months. This trend makes sense since ridership would have been closely correlated to the outdoor weather conditions (i.e. more customers using the service when it is warm and sunny). To get a better understanding of the difference in the total trip duration between casual riders and annual members, we add an additional condition in the function to group by the member_casual variable.

```r
monthly_total_trip_duration <- complete_divvy_tripdata %>%
  group_by(month = lubridate::floor_date(started_at, "month"), member_casual) %>%
  summarize(total_trip_duration = sum(trip_duration))

view(monthly_total_trip_duration)
```
